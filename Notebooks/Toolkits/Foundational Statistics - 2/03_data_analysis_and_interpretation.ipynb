## Why Interpret Data?

Data interpretation is not just a statistical process; it‚Äôs a strategic tool that helps **businesses make informed decisions**. It‚Äôs also a fundamental step in preparing for the exciting world of machine learning.

Why do businesses like ‚ÄòSodaPop‚Äô, a soft drink company, need to delve into data interpretation? ü§î The answer lies in the need to answer crucial business questions, such as:

- ‚ÄúWhich of our products do customers prefer?‚Äù or
- ‚ÄúWhat are the main factors influencing a customer‚Äôs decision to choose between ‚ÄòSodaPop Classic‚Äô and ‚ÄòSodaPop Zero‚Äô?‚Äù

In the case of ‚ÄòSodaPop‚Äô, they face a challenge. They have gathered customer survey data but are unsure how to interpret it to gain valuable insights about customer preferences. This is where data interpretation comes in.

## Data Interpretation: A Critical Precursor to Machine Learning

For instance, some customers might prefer ‚ÄúSodaPop Classic‚Äù due to its taste, while others might select ‚ÄúSodaPop Zero‚Äù for its lower calorie content. The interpretation process helps SodaPop uncover these consumer preferences and behaviors, setting the stage for predictive modeling and algorithmic decision-making in machine learning.

Data interpretation comes in two general flavors, both of which are intimately linked to machine learning:

1. **Quantitative**, focusing on numbers and statistics.
2. **Qualitative**, emphasizing descriptions and interpretations.

Before embarking on the data interpretation journey, SodaPop needs to decide on the data measurement **scale**. They might categorize data by product (a **nominal** scale), sort it by customer satisfaction level (an **ordinal** scale), analyze the distribution of customer ages (an **interval** scale) or analyze the ratio of customers who choose ‚ÄúSodaPop Classic‚Äù compared to ‚ÄúSodaPop Zero‚Äù (a **ratio** scale).

With the measurement scale selected, SodaPop can now choose the interpretation process that best suits their data needs. ## Data Interpretation in Practice: The SodaPop Example

Imagine we have survey data from the SodaPop company. This data includes columns for 'Product' (the product the customer chose), 'Age' (the customer's age), 'Satisfaction' (the level of customer satisfaction), and 'CalorieConcern' (whether the customer is conscious about calorie intake).

```python
# @title #### Generate Example Survey SodaPop

import pandas as pd
import numpy as np

num_customers = 500

products = np.random.choice(['SodaPop Classic', 'SodaPop Zero'], num_customers)

ages = np.random.randint(18, 70, num_customers)

satisfaction_levels = ['Very Unsatisfied', 'Unsatisfied', 'Neutral', 'Satisfied', 'Very Satisfied']
satisfaction = np.random.choice(satisfaction_levels, num_customers, p=[0.1, 0.1, 0.3, 0.3, 0.2])

calorie_concern = np.random.choice([0, 1], num_customers, p=[0.5, 0.5])  # 0 = Tidak, 1 = Ya

df = pd.DataFrame({
    'Product': products,
    'Age': ages,
    'Satisfaction': satisfaction,
    'CalorieConcern': calorie_concern
})

df.head()
```

> **Note**: Do note that this generated data serves as an example and might not fully reflect the actual situation a soft drink company may encounter.

### Applying Data Interpretation and Preparing for Machine Learning

#### 1. Nominal Data

For instance, we want to know the distribution of product preferences between "SodaPop Classic" and "SodaPop Zero". We can create a bar chart to display the number of customers who choose each product. This kind of categorical data is often used in machine learning for classification tasks.

```python
# @title #### Nominal Charts

import warnings
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    ax = sns.countplot(x='Product', data=df, palette='viridis', order=df['Product'].value_counts().index)

ax.set_title('Product Preference of Customers', fontsize=16, fontweight='bold')
ax.set_xlabel('Product', fontsize=12)
ax.set_ylabel('Number of Customers', fontsize=12)

sns.despine(ax=ax, top=True, right=True)

plt.xticks(rotation=45)
plt.show()
```

The chart is used to determine which 'Product' is favored by most customers, as indicated by the height of the bars representing each 'Product'.

If, for example, the chart shows that 'SodaPop Classic' has a significantly higher bar (more counts), it means more customers prefer 'SodaPop Classic' over other products.

- Based on this information, the company might **decide to allocate more resources to the production and marketing** of 'SodaPop Classic' to meet customer demand and maximize sales.
- Conversely, if a product has few customers, the company might **reassess its production** or even consider **discontinuing it**.

#### 2. Ordinal Data

Our data includes customer satisfaction levels, ranging from "Very dissatisfied" to "Very satisfied". A bar chart can showcase the number of customers at each satisfaction level. Ordinal data like this can be transformed and used in machine learning to predict customer satisfaction or churn.

```python
import warnings
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    satisfaction_counts = df['Satisfaction'].value_counts().sort_index()
    ax = sns.barplot(x=satisfaction_counts.index, y=satisfaction_counts.values, palette='viridis')

ax.set_title('Customer Satisfaction Levels', fontsize=16, fontweight='bold')
ax.set_xlabel('Satisfaction Level', fontsize=12)
ax.set_ylabel('Number of Customers', fontsize=12)

sns.despine(ax=ax, top=True, right=True)

plt.xticks(rotation=45)

plt.show()
```

For instance, a high bar for 'Very Satisfied' signifies a significant proportion of customers being highly satisfied with the product/service. On the flip side, a high bar for 'Very Unsatisfied' suggests a substantial number of unhappy customers.

- This data is critical for the company's strategic decisions. For instance, if many customers are unsatisfied, it's crucial to investigate reasons and take steps for improvement.
- Conversely, if most customers are satisfied, the company might want to keep their current practices and focus on converting the 'Neutral' customers into more satisfied ones.

#### 3. Interval Data

Creating a histogram of customer ages allows us to see the distribution of customer ages. Interval data is a fundamental component of many machine learning algorithms, providing continuous input variables.

```python
# @title #### Interval Charts

import seaborn as sns
import matplotlib.pyplot as plt

# Atur style plot
sns.set_style("whitegrid")

# Buat plot
ax = sns.histplot(df['Age'].dropna(), bins=10, kde=True, color='skyblue')

# Judul dan label dengan font yang besar dan tebal
ax.set_title('Age Distribution of Customers', fontsize=16, fontweight='bold')
ax.set_xlabel('Age', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)

# Hapus garis batas atas dan kanan
sns.despine(ax=ax, top=True, right=True)

plt.show()
```

The argument 'bins=10' evenly divides the age data into 10 intervals. Each bar's height corresponds to the customer count in that age range. A higher bar in a specific age group indicates a larger concentration of customers within that age range.

For instance, a tallest bar in the 30-40 age range suggests most customers fall in this bracket. In contrast, a lower bar suggests a smaller customer base within that age group. This information assists the company in tailoring their marketing strategies and product development based on the age group concentrations.

- If the majority of customers are young, the company may focus on products appealing to a younger demographic and leverage marketing channels that effectively reach this group.
- Conversely, if a certain age group is underrepresented, the company may devise strategies to attract more customers from this demographic, or shift focus to more prevalent age groups.

#### 4. Ratio Data

We might be interested in the ratio of customers who choose "SodaPop Classic" compared to "SodaPop Zero". By calculating and displaying this ratio, we can gain insights into customer preferences. Ratio data like this is often used in machine learning for tasks like regression or clustering.

$$\text{{Ratio}} = \frac{{\text{{Classic\ amount }}}}{{\text{{Zeros\ amount}}}}$$

```python
# @title #### Ratio

classic_count = df[df['Product'] == 'SodaPop Classic'].shape[0]
zero_count = df[df['Product'] == 'SodaPop Zero'].shape[0]
ratio = classic_count / zero_count

print(f'The ratio of customers who chose SodaPop Classic to SodaPop Zero is {ratio:.2f}')
```

After selecting a measurement scale and performing data interpretation, ‚ÄúSodaPop‚Äù may want to perform a trend analysis to see how customer preferences change over time, or they may want to perform a correlation analysis to see the relationship between certain variables (for example, age and product preferences).

An example of correlation analysis can be done with library: `seaborn`

```python
# @title #### Correlation

import seaborn as sns
import matplotlib.pyplot as plt

# Mapping levels of satisfaction
satisfaction_mapping = {
    'Sangat tidak puas': 1,
    'Tidak puas': 2,
    'Netral': 3,
    'Puas': 4,
    'Sangat puas': 5
}
df['Satisfaction'] = df['Satisfaction'].replace(satisfaction_mapping)

# Atur style plot
sns.set_style("whitegrid")

# Buat correlation matrix
corr_matrix = df[['Age', 'Satisfaction']].corr()

# Buat heatmap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
ax = sns.heatmap(corr_matrix, annot=True, cmap=cmap, center=0,
                 annot_kws={"size": 14}, linewidths=.5)

# Judul dengan font yang besar dan tebal
ax.set_title('Correlation between Age and Satisfaction', fontsize=16, fontweight='bold')

# Label axis dengan font size 12
plt.xlabel('Features', fontsize=12)
plt.ylabel('Features', fontsize=12)

plt.show()
```

The heatmap showcases the correlation coefficient between 'Age' and 'Satisfaction', reflecting the strength and direction of their relationship. The coefficient ranges from -1 to 1, with a value near 1 indicating a strong positive correlation, near -1 indicating a strong negative correlation, and around 0 suggesting no clear correlation.

The colormap of the heatmap signifies the correlation coefficient, with darker colors generally representing stronger correlations (either positive or negative), and lighter colors indicating weaker correlations.

This data aids in understanding how customer characteristics correlate with their satisfaction. For instance, a strong positive correlation between age and satisfaction implies that older customers are typically more satisfied. Thus, the company might wish to investigate this and develop strategies to improve satisfaction among younger customers.

## Advanced Analysis Using the "Titanic" Dataset from Hugging Face

For more advanced analysis, we can use a larger and more complex dataset from Hugging Face. Hugging Face has many public datasets available, but for this example, we will use the "Titanic" dataset.

The ‚ÄúTitanic‚Äù dataset has information about Titanic passengers, including age, gender, passenger class, and whether they survived or not. Let's load this dataset and do some analysis:

```python
import pandas as pd

# URL dari dataset Titanic di GitHub
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Muat dataset
df = pd.read_csv(url)
df.head()
```

Keep in mind that before performing this analysis, we may need to do some data pre-processing. Data pre-processing is an important step before performing any analysis. In the Titanic dataset, some columns may have missing values (NaN) that we need to handle.

Here are some common strategies for dealing with missing values:

- **Deletion**: You can choose to delete rows or columns that have missing values. However, this may result in the loss of important information.

```python
# Deletes rows with missing values
df_cleaned = df.dropna()

# Removes columns with missing values
df_cleaned = df.dropna(axis=1)
df_cleaned.head()
```

- **Imputasi**: Anda dapat memilih untuk mengisi nilai yang hilang dengan nilai lain, seperti mean, median, atau mode dari kolom.

```python
# Only select numeric columns
df_numeric = df.select_dtypes(include=[np.number])

# Fill in the missing values with the median of each column
df_filled_numeric = df_numeric.fillna(df_numeric.median())

# Replace numeric columns in df with df_filled_numeric
df[df_filled_numeric.columns] = df_filled_numeric

df.head()
```

- **Prediction**: You can use more sophisticated methods, such as regression or other model-based methods, to predict missing values.

For the Titanic dataset, the 'age' column has some missing values. Let's fill these values with the median:

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df.head()
```

Now, we can continue with our **analysis** and **visualization** with the cleaned and processed data:

#### 1. Nominal Data

Nominal data is data that is categorized without a particular order or hierarchy. This is a type of qualitative data and is usually used to label variables without quantitative values. In the Titanic dataset, the columns 'sex' and 'embarked' (departure location) are examples of nominal data.

- Column 'sex': There are two categories in this column, 'male' and 'female'. These are purely labels and there is no order or hierarchy in this data.

- Column 'embarked': There are three categories in this column, 'S', 'C', and 'Q', representing Southampton, Cherbourg, and Queenstown respectively. Like 'sex', this is also purely a label and there is no order or hierarchy.

Nominal data analysis usually involves calculating frequencies and percentages. For example, in the context of the Titanic dataset, we might be interested in seeing the distribution of passengers by gender or departure location. Using visualizations such as count plots can help us understand the proportion of male and female passengers, as well as where passengers are more likely to depart.

What is nominal data analysis used for? This knowledge may be useful for further analysis or predictive models. For example, if we find that female passengers have a higher survival rate, our model might take this into account. Similarly, if passengers from Cherbourg had a higher survival rate, our model might also take this into account.

So, basically, this analysis helps us understand the basic characteristics of passengers and may provide insight into what factors might influence their survival rates.

```python
# @title #### Nominal Charts

import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Warna untuk pie chart
colors = ['#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

# Pie chart untuk 'Sex'
sex_counts = df['Sex'].value_counts()
axes[0].pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
axes[0].set_title('Passenger Gender Distribution', fontsize=12)

# Pie chart untuk 'Embarked'
embarked_counts = df['Embarked'].value_counts()
axes[1].pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)
axes[1].set_title('Passenger Embarkment Distribution', fontsize=12)

# Equal aspect ratio ensures that pie is drawn as a circle
axes[0].axis('equal')
axes[1].axis('equal')

plt.tight_layout()
plt.show()
```

#### 2. Ordinal Data

Ordinal data is a categorical data type that has a certain order or hierarchy. In the Titanic dataset, the 'pclass' column (passenger class) is an example of ordinal data.

- Column 'pclass': There are three categories in this column, namely 1, 2, and 3, which respectively represent Class 1, Class 2, and Class 3 on the Titanic. This category has an order, meaning that Class 1 is higher than Class 2 and 3, and Class 2 is higher than Class 3.

Just like with nominal data, we can perform frequency and percentage analysis on ordinal data. However, because ordinal data also has order, we can perform some additional analyzes that cannot be done on nominal data.

For example, we might be interested to see how survival rates differ between different classes of passengers. We may find that passengers in Class 1 have a higher survival rate compared to passengers in Classes 2 and 3. This may help us understand whether passenger 'class' is an important factor in survival rates.

Here is how we can visualize the passenger class distribution:

```python
# @title #### Ordinal Charts

import warnings
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    ax = sns.countplot(x='Pclass', data=df, order=[1, 2, 3], palette='viridis')

plt.title('Passenger Class Distribution', fontsize=16, fontweight='bold')

ax.set_xlabel('Passenger Class', fontsize=12)
ax.set_ylabel('Count', fontsize=12)

sns.despine(ax=ax, top=True, right=True)

plt.show()
```

This knowledge is important for further analysis and predictive models. For example, if we find that passengers in Class 1 have a higher survival rate, our model might take this into account when making predictions.

#### 3. Interval Data

Interval data is a numeric data type that has a fixed order and distance between values, but does not have a true or absolute starting point of zero. In the Titanic dataset, the 'age' column is an example of interval data.

- Column 'age': The values in this column represent the passenger's age in years. These ages are ordered (a 30 year old passenger is older than a 20 year old passenger) and have a fixed distance between values (the distance between 20 and 30 year olds is the same as the distance between 30 and 40 year olds).

With interval data, we can perform all kinds of mathematical operations (addition, subtraction, division, and multiplication) and descriptive statistics such as mean, median, and mode. For example, we might be interested in knowing the average age of passengers or the age distribution of passengers.

Here is how we can visualize the age distribution of passengers:

```python
# @title #### Interval Charts

import seaborn as sns
import matplotlib.pyplot as plt

# Atur style plot
sns.set(style="whitegrid")

# Membuat histogram untuk 'Age'
sns.histplot(data=df, x='Age', bins=30, color='skyblue', kde=True)

# Judul plot dengan font yang besar dan tebal
plt.title('Age Distribution of Passengers', fontsize=16, fontweight='bold')

# Label sumbu x dan y dengan font yang lebih besar
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Hapus garis batas atas dan kanan
sns.despine()

plt.show()
```

This analysis can help us understand the age distribution of passengers and may provide insight into how age might influence survival rates. For example, if we find that younger passengers tend to have higher survival rates, our model might take this into account when making predictions.

#### 4. Ratio Data

Ratio data is a numeric data type that has all the characteristics of interval data (fixed order and distance between values), plus has a true or absolute zero starting point. In the Titanic dataset, the 'fare' column is an example of ratio data.

- 'fare' column: The values in this column represent the fare paid by the passenger. These rates have a sequence (a passenger paying 30 has a higher fare than a passenger paying 20) and have a fixed distance between values (the distance between 20 and 30 is the same as the distance between 30 and 40). Additionally, a starting point of zero in this data is meaningful, i.e. the fare can be zero (the passenger pays nothing).

With ratio data, we can perform all kinds of mathematical operations as well as descriptive statistics such as mean, median, and mode. For example, we might be interested in knowing the average fare paid by passengers or the distribution of passenger fares.

Here is how we can visualize the distribution of passenger fares:

```python
# @title #### Ratio Charts

import seaborn as sns
import matplotlib.pyplot as plt

# Atur style plot
sns.set_style("whitegrid")

# Buat plot
ax = sns.histplot(df['Fare'].dropna(), bins=30, color='skyblue', kde=True)

# Judul dan label dengan font yang besar dan tebal
ax.set_title('Fare Distribution of Passengers', fontsize=16, fontweight='bold')
ax.set_xlabel('Fare', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)

# Hapus garis batas atas dan kanan
sns.despine(ax=ax, top=True, right=True)

plt.show()
```

This analysis can help us understand the distribution of passenger fares and may provide insight into how fares might affect survival rates. For example, if we find that passengers who pay higher fares tend to have higher survival rates, our model might take this into account when making predictions.
